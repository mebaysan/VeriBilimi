```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt # görselleştirme kütüphanesi
```

```python
data = pd.read_csv('diabetes.csv')
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
df.describe().T # betimleyici istatistikleri verir
```

```python
df.info() # veri tiplerini verir
```

```python
df.isna().any() # değişkenlerin gözlemlerinde hiç NaN değer var mı?
```

```python
df.notna().any() # değişkenlerin gözlemlerinde hiç NaN değer olmayan var mı?
```

```python
df.notna().all() # değişkenlerden hangisi tamamen NaN değil
```

```python
df.isna().sum() # değişkenlerde NaN olan gözlemlerin sayısı
```

```python
plt.scatter(
    x = 'total_bill', # x ekseni
    y = 'tip', # y ekseni
    data = df, # hangi veri seti
    c = 'size', # hangi değişkene göre sınıflandıracağız
    s = 289, # marker boyutu
    alpha = 0.2 # marker transparanlığı
)
plt.colorbar(); # hangi renk hangi değere karşılık geliyor için sağ tarafa paleti koyar
#plt.xlabel('Toplam Hesap')
#plt.ylabel('Bahşiş')
plt.title('Toplam Hesap ve Bahşiş Arasındaki İlişki')
plt.show()
```

```python
fig, ax = plt.subplots(2,1) # fig -> plot'a ekleyeceğimiz her şey burada tutulur, ax -> eksenler üzerindeki veri

plt.show()
```

```python
fig, ax = plt.subplots()

plt.show()
```

```python
fig, ax = plt.subplots() # subplots içine bir şey vermezsek 1'e 1'lik (yani tek grafik) oluşturur. İlk parametre satır, ikinci parametre sütun
ax.scatter(x='total_bill',y='tip',c='size',data=df,cmap='PuBu') # cmap için bkz. -> https://matplotlib.org/tutorials/colors/colormaps.html
ax.set_xlabel('Toplam Hesap')
ax.set_ylabel('Bahşiş')
ax.set_title('Toplam Hesaba Göre Verilen Bahşiş Arasındaki İlişki')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(x='tip',data=df,bins=5)
ax.set_xlabel('Bahşiş')
plt.show()
```
