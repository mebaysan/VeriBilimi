```python
def hesap_sinifla(data):
    if data <=20:
        print(f'Düşük Hesap: {data}')
    elif data >= 35:
        print(f'Yüksek Hesap: {data}')
    else: 
        print(f'Normal Hesap: {data}')

total_bill.apply(hesap_sinifla) # appyl methodu ile kendi yazdığımız (veya hazır da olabilir) fonksiyonları seri üzerinde uygulayabiliriz
```
