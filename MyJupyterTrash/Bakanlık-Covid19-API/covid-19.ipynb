```python
import pandas as pd
import requests
```

```python
raw = requests.get("https://covid19.saglik.gov.tr/covid19api?getir=liste") # istek atıp dönen veriyi değişkene atıyoruz
raw_json = raw.json() # dönen veri içerisindeki json'u alıyoruz
DF = pd.DataFrame(raw_json) # dönen json'dan bir DataFrame oluşturuyoruz
```

```python
cols = DF.columns.tolist() # kolonları (değişkenleri) bir listeye atıyoruz
new_cols = [] 
for col_name in cols:
    if 'gunluk' in col_name or 'toplam' in col_name or 'tarih' in col_name:
        new_cols.append(col_name) # istediğimiz kolonlardan bir liste oluşturuyoruz
```

```python
DF = DF[new_cols] # yeni kolonları seçip bir DF oluşturuyoruz
```

```python
for col_name in DF.loc[:, 'gunluk_test':].columns: # gunluk_test (dahil) değişkeninden sonraki değişkenleri gez
    DF[f'{col_name}'] = DF[f'{col_name}'].str.replace('.', '').astype('string') # her birini önce '.' dan arındır ve birleştir. Sonra da stringe çevir
    DF[f'{col_name}'] = pd.to_numeric(DF[f'{col_name}']) # her değişkeni sayısala çevir
```

```python
DF.fillna(0, inplace=True) # boş olan değerleri 0 ile doldur
```

```python
for col_name in DF.loc[:, 'gunluk_test':].columns: # gunluk_test (dahil) değişkeninden sonraki değişkenleri gez
    DF[f'{col_name}'] = DF[f'{col_name}'].astype('int') # her birini int'e çevir
```

```python
DF['tarih'] = pd.to_datetime(DF['tarih'], format='%d.%m.%Y') # tarih formatına göre elimdeki Date verisini pandas date tipine çevir
```
